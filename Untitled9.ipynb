````python
import textwrap
import re
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# ============================================
#   SIMPLE TEXT PARSER
# ============================================
def parse_events_from_text(text):
    """Parse events from structured text format."""
    events = []

    # Split by EVENT markers
    event_blocks = re.split(r'EVENT \d+:', text)

    for block in event_blocks[1:]:  # Skip first empty split
        event = {
            "speaker": "",
            "utterance": "",
            "action_type": "",
            "sentiment": "",
            "relevance": ""
        }

        # Extract each field
        speaker_match = re.search(r'Speaker:\s*(.+)', block)
        if speaker_match:
            event["speaker"] = speaker_match.group(1).strip()

        utterance_match = re.search(r'Utterance:\s*"([^"]+)"', block)
        if utterance_match:
            event["utterance"] = utterance_match.group(1).strip()

        action_match = re.search(r'Action:\s*(.+)', block)
        if action_match:
            event["action_type"] = action_match.group(1).strip()

        sentiment_match = re.search(r'Sentiment:\s*(.+)', block)
        if sentiment_match:
            event["sentiment"] = sentiment_match.group(1).strip()

        relevance_match = re.search(r'Relevance:\s*(.+)', block)
        if relevance_match:
            event["relevance"] = relevance_match.group(1).strip()

        if event["utterance"]:  # Only add if we got at least an utterance
            events.append(event)

    return {"events": events}

def parse_causal_from_text(text):
    """Parse causal explanation from structured text format."""
    result = {
        "causal_explanation": "",
        "causes": []
    }

    # Extract explanation
    explanation_match = re.search(r'EXPLANATION:\s*(.+?)(?=CAUSE \d+:|$)', text, re.DOTALL)
    if explanation_match:
        result["causal_explanation"] = explanation_match.group(1).strip()

    # Extract causes
    cause_blocks = re.split(r'CAUSE \d+:', text)

    for block in cause_blocks[1:]:
        cause = {
            "cause": "",
            "mechanism": "",
            "evidence": []
        }

        cause_match = re.search(r'What:\s*(.+)', block)
        if cause_match:
            cause["cause"] = cause_match.group(1).strip()

        mechanism_match = re.search(r'How:\s*(.+)', block)
        if mechanism_match:
            cause["mechanism"] = mechanism_match.group(1).strip()

        # Extract evidence items
        evidence_matches = re.findall(r'- (.+)', block)
        cause["evidence"] = [e.strip() for e in evidence_matches]

        if cause["cause"]:
            result["causes"].append(cause)

    return result

# ============================================
#   5. EVENT EXTRACTION PROMPT (TEXT FORMAT)
# ============================================
event_prompt = ChatPromptTemplate.from_messages([
    ("system", "You extract events from conversation transcripts. Be precise and use exact quotes."),
    ("user", """Extract events relevant to: {query}

Context:
{context}

Format your response EXACTLY like this:

EVENT 1:
Speaker: A
Utterance: "exact quote from context"
Action: describing what they did
Sentiment: positive/negative/neutral
Relevance: high/medium/low

EVENT 2:
Speaker: B
Utterance: "another exact quote"
Action: describing action
Sentiment: negative
Relevance: high

Use ONLY quotes from the context. Extract 3-5 most relevant events.""")
])

event_chain = event_prompt | llm | StrOutputParser()

# ============================================
#   6. CAUSAL EXPLANATION PROMPT (TEXT FORMAT)
# ============================================
causal_prompt = ChatPromptTemplate.from_messages([
    ("system", "You analyze events and explain causal relationships clearly."),
    ("user", """Analyze the cause for: {query}

Events that occurred:
{events_text}

Format your response EXACTLY like this:

EXPLANATION:
Write 2-3 sentences explaining what happened and why.

CAUSE 1:
What: Brief description of the cause
How: Explain the mechanism/how it led to the outcome
Evidence:
- "quote or event 1"
- "quote or event 2"

CAUSE 2:
What: Another contributing factor
How: How this factor contributed
Evidence:
- "supporting quote"

Identify 2-3 main causes.""")
])

causal_chain = causal_prompt | llm | StrOutputParser()

# ============================================
#   PIPELINE
# ============================================
def causal_explanation_pipeline(query):
    try:
        # Retrieval
        retrieved_docs = retriever.invoke(query)
        context = "\n\n".join([doc.page_content for doc in retrieved_docs])

        # Step 1: Event extraction
        print(f"🔍 Extracting events for: {query}")
        event_output = event_chain.invoke({
            "context": context,
            "query": query
        })

        print(f"\n📄 Raw event output:\n{event_output}\n")
        print("-" * 60)

        events_data = parse_events_from_text(event_output)

        # Create text summary of events for causal step
        events_text = ""
        for i, event in enumerate(events_data["events"], 1):
            events_text += f"{i}. {event['speaker']}: \"{event['utterance']}\" ({event['sentiment']})\n"

        # Step 2: Causal explanation
        print(f"\n🔬 Generating causal explanation...")
        causal_output = causal_chain.invoke({
            "events_text": events_text,
            "query": query
        })

        print(f"\n📄 Raw causal output:\n{causal_output}\n")
        print("-" * 60)

        causal_data = parse_causal_from_text(causal_output)

        return {
            "retrieved_docs": retrieved_docs,
            "events": events_data,
            "causal": causal_data,
            "raw_event_output": event_output,
            "raw_causal_output": causal_output,
            "success": True
        }

    except Exception as e:
        print(f"❌ Pipeline error: {e}")
        import traceback
        traceback.print_exc()
        return {
            "error": str(e),
            "events": {"events": []},
            "causal": {"causal_explanation": "Pipeline failed", "causes": []},
            "success": False
        }

# ============================================
#   8. TEST EXAMPLES
# ============================================
queries = [
    "What escalations happened during the meet?",
    "Why did the conflict arise?",
    "Why did the client get annoyed?",
    "Explain the cause of disagreement."
]

from pprint import pprint

for q in queries:
    print("\n" + "="*70)
    print(f"🔹 QUERY: {q}")
    print("="*70)

    result = causal_explanation_pipeline(q)

    if not result.get("success", False):
        print(f"\n⚠️  ERROR: {result.get('error', 'Unknown error')}")
        continue

    print("\n" + "✅ STRUCTURED EVENTS:")
    pprint(result["events"], width=120)

    print("\n" + "✅ CAUSAL ANALYSIS:")
    pprint(result["causal"], width=120)

    print("\n" + "="*70)
````
